```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,13)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X-axies")
plt.ylabel("Y-axies")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,y)
plt.subplot(2,2,4)
plt.plot(y==2,x)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis1 = fig.add_axes([0.3,0.3,0.4,0.4])
axis1.plot(x,y)
```
